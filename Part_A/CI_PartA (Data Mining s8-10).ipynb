#### Open site and copy all csv files from there

```python
array_csv=[]#*len(subject)
for s in range(len(subject)):
    
    address = f"{perent_address}s{subject[s]}/"
    print(f'Address: {address} -> S:{subject[s]}')
    path = "//table/tbody/tr/td/a"
    site = requests.get(address).content
    all_links = html.fromstring(site).xpath(path)    
    links = []
    
    for i in range(len(all_links)):
        #find csv files
        if all_links[i].text == 'csv':
            csv = all_links[i].attrib['href']            
            #add to array
            links.append(f"https://www.uni-mannheim.de{csv}")
            
    array_csv.append(links)
    print(f"ARAAAAY:{array_csv[s]}")
    print("________________________")
```

### Merge sensors with chosen activities

```python
activities = ['climbingdown', 'running', 'jumping', 'lying']
sensors = ['acc', 'lig','mic']
body_part = 'thigh'
```

```python
import pandas as pd
import os

sets = []#*len(subject)
#print(len(subject))
for s in range(len(subject)):
    activSensors = {}
    path = f'/Users/miror/OneDrive - Ye9pfoBh9P/1_VUM_STUDY/3 year/01 Computational Intelligence/Assignment/Subject_{subject[s]}'

    for (dirpath, dirnames, filenames) in os.walk(path):
        for filename in filenames:
            file=os.path.join(dirpath, filename)
            for a in activities:
                for i in sensors:
                    if i in file:
                        if '.csv' in file and body_part in file:
                            activity = file.split('_')[-2]
                            if activity == a:
                                temp=pd.read_csv(file)
                                temp.columns = [f'{i}_{str(col)}' for col in temp.columns] 
                                if activity in activSensors:
                                    activSensors[activity][i] = temp
                                else:
                                    activSensors[activity] = { i : temp }
    print(activSensors)
    print("-------------------------------")
    sets.append(activSensors)
```

### Merge to one dimensional array

```python
concat_activities=[]
i=0
for activSensors in sets:
    table = []
    print(f'_______________-Subject {subject[i]}-______________')
    for activity in activSensors:
        print(f'\n \n{activity}')
        
        temp_row = pd.concat([activSensors[activity]['acc'], activSensors[activity]['lig'], activSensors[activity]['mic']], axis=1)
        print(temp_row)
        #add activity collumn
        temp_row['activity']=activity
        temp_row['subject']=subject[i]
        table.append(temp_row)
        
    concat_activities.append(table)
    i+=1
```

```python
tables = []
for ca in concat_activities:
    dt = pd.DataFrame()
    dt = pd.concat(ca)
    tables.append(dt)
```
